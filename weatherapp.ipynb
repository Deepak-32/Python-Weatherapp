```python
import requests
from datetime import datetime

def fetch_current_weather(city, api_key):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)
    data = response.json()
    return data

def fetch_5_day_forecast(city, api_key):
    url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)
    data = response.json()
    return data

def display_current_weather(weather_data):
    print("Current Weather Conditions:")
    print(f"Temperature: {weather_data['main']['temp']}°C")
    print(f"Weather Description: {weather_data['weather'][0]['description']}")
    print(f"Humidity: {weather_data['main']['humidity']}%")
    print(f"Wind Speed: {weather_data['wind']['speed']} m/s")
    sunrise = datetime.fromtimestamp(weather_data['sys']['sunrise']).strftime('%Y-%m-%d %H:%M:%S')
    sunset = datetime.fromtimestamp(weather_data['sys']['sunset']).strftime('%Y-%m-%d %H:%M:%S')
    print(f"Sunrise Time: {sunrise}")
    print(f"Sunset Time: {sunset}")

def display_5_day_forecast(forecast_data):
    print("\n5-Day Weather Forecast:")
    for item in forecast_data['list']:
        date = datetime.fromtimestamp(item['dt']).strftime('%Y-%m-%d')
        print(f"\nDate: {date}")
        print(f"Temperature: {item['main']['temp']}°C")
        print(f"Weather Description: {item['weather'][0]['description']}")

def main():
    city = input("Enter city name: ")
    api_key = "7006d7d1e17b00fdde11c48019b271f1"  # Replace this with your OpenWeatherMap API key
    current_weather = fetch_current_weather(city, api_key)
    if current_weather['cod'] == 200:
        display_current_weather(current_weather)
    else:
        print("City not found. Please try again.")
        return

    forecast_data = fetch_5_day_forecast(city, api_key)
    if forecast_data['cod'] == '200':
        display_5_day_forecast(forecast_data)
    else:
        print("Forecast data not available.")

if __name__ == "__main__":
    main()
```
